#**Bivariate Analysis**

```python
#Create a scatterplot for Age and Duration of Pitch (Also check product is taken or not by using hue)
sns.scatterplot(x = 'Age', y = 'DurationOfPitch', data = df, hue = 'ProdTaken', palette = 'coolwarm')
plt.title("Scatterplot for Age and Duration of Pitch")
plt.xlabel('Age')
plt.ylabel('Duration of Pitch')
plt.show()
```

**Insight**-The scatterplot suggests that neither age nor pitch duration strongly influences whether the product is taken. The data points are fairly evenly distributed across both variables, with a majority of pitches being relatively short in duration.

```python
#Stacked Bar Chart- Maritial Status and Product Taken
crosstab = pd.crosstab(df['MaritalStatus'], df['ProdTaken'], normalize = 'index')
crosstab
```

```python
crosstab.plot(kind = 'bar', stacked = True, color = ['skyblue', 'lightcoral'])
plt.title("Stacked Bar Chart for Marital Status and Product Taken")
plt.xlabel('Maritial Status')
plt.ylabel('Percentage')
plt.show()
```

**Insight**-Single individuals are more likely to take the product compared to those in other marital status categories.

```python
#Creat a Bar Chart for Product Pitch and Satisfaction Score
sns.barplot(x = 'ProductPitched', y = 'PitchSatisfactionScore', data = df, palette = 'coolwarm' )
plt.title("Bar Chart for Product Pitched and Pitched Satisfaction Score")
plt.xlabel('Product Pitched')
plt.ylabel('Pitched Satisfaction Score')
plt.show()
```

**Insight**-The "Super Deluxe" and "King" products are the highest-rated in terms of customer satisfaction, with "Super Deluxe" leading slightly.

```python
#Lineplot-No. of Follow up and Pitch Satisfaction
sns.lineplot(x = 'NumberOfFollowups', y = 'PitchSatisfactionScore', data = df, marker = 'o', color = 'green')
plt.title("Line Cahrt for No. of Followups and Pitch Satifaction Score")
plt.xlabel('No. of Followups')
plt.ylabel('Pitch Satifaction Score')
plt.show()
```

**Insight**-The highest satisfaction scores are observed when there are 3 to 4 follow-ups.

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
#For Numerical features use Standard Scaler (Scaling)
#For categorical Features use One Hot Encoding
#Use Column Transformer to do both function in one go.
```

```python
num_transformer = StandardScaler()
oh_transformer = OneHotEncoder()
```

```python
preprocessor = ColumnTransformer(
    [
        ("Encoder", oh_transformer, cat_features),
        ("Scaler", num_transformer, num_features)
    ]
)
```

```python
x_train = preprocessor.fit_transform(x_train)
```
